```python
# imports
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

from frb.associate import chance
```

```python
sns.set_theme()
sns.set_style('whitegrid')
sns.set_context('talk')
```

# Compare Driver and Bloom

```python
m_r = np.linspace(13., 30., 1000)
```

```python
bloom = chance.bloom_sigma(m_r)
```

```python
driver = chance.driver_sigma(m_r)
```

## Plot

```python
fig = plt.figure(figsize=(5,5))
ax = sns.lineplot(x=m_r, y=driver, label='Driver', ls=':')
ax.set_yscale("log", nonpositive='clip')
#
ax.plot(m_r, bloom, label='Bloom', ls='--')
ax.legend(loc='upper left')
```
